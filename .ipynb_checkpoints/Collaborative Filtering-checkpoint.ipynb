```python
ratings = pd.read_csv("ml-latest-small/ratings.csv")
ratings.head(5)
```

```python
ratings.info()
```

```python
#dropping the timestamp column
ratings2 = ratings.drop(['timestamp'], axis=1)
```

```python
ratings2.head()
```

```python
ratings2['movieId'].nunique()
```

```python
#selecting random user 100
userId = 100
inputMovies = ratings2[ratings2['userId']== userId]
inputMovies.head(10)
```

```python
len(inputMovies)
```

```python
#get similar users who have rated similar movies as user 100
similarUsers = ratings2[ratings2['movieId'].isin(inputMovies['movieId'].tolist())]
similarUsers
```

```python
#reshaping the data to table based on column values
user_item_ptable= ratings2.pivot(index='userId', columns='movieId', values='rating')
user_item_ptable.head(5)
```
